```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

plt.style.use('seaborn')
sns.set(style="white", context="talk")
```

```python
rs = np.random.RandomState(8)
rs
```

```python
x = np.array(list("ABCDEFGHIJ"))
x
```

```python
y1 = np.arange(1, 11)
y1
```

```python
sns.barplot(x=x, y=y1, palette="rocket").axhline(0, color="k", clip_on=False)
sns.despine(bottom=True)
```

```python
y2 = y1 - 5.5
y2
```

```python
sns.barplot(x=x, y=y2, palette="vlag").axhline(0, color="k", clip_on=False)
sns.despine(bottom=True)
```

```python
y3 = rs.choice(y1, len(y1), replace=False)
y3
```

```python
sns.barplot(x=x, y=y3, palette="deep").axhline(0, color="k", clip_on=False)
sns.despine(bottom=True)
```

```python
# Set up the matplotlib figure
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 5), sharex=True)

sns.barplot(x=x, y=y1, palette="rocket", ax=ax3)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Sequential")

sns.barplot(x=x, y=y2, palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Diverging")

sns.barplot(x=x, y=y3, palette="deep", ax=ax1)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Qualitative")


# Finalize the plot
sns.despine(bottom=True)
plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```
