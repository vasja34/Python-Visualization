```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

plt.style.use('seaborn')
sns.set(style="whitegrid")
```

```python
# Make an example dataset with y ~ x
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
```

```python
sns.scatterplot(x=x, y=y, color="g")
```

```python
sns.regplot(x=x, y=y, color="g")
```

```python
# lowess: locally weighted robust scatterplot smoothing
sns.regplot(x=x, y=y, color="g", lowess=True)
```

```python
# Plot the residuals after fitting a linear model
sns.residplot(x, y, lowess=True, color="g")
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2)
fig.set_size_inches(15, 8)
sns.scatterplot(x, y, color="g", ax=axes[0][0])
sns.regplot(x, y,color="g", ax=axes[0][1])
sns.regplot(x, y, lowess=True, color="g", ax=axes[1][0])
sns.residplot(x, y, lowess=True, color="g", ax=axes[1][1])
```
