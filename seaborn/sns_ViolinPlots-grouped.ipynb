## Grouped violinplots with split violins
* https://seaborn.pydata.org/examples/grouped_violinplots.html

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

plt.style.use('seaborn')
sns.set(style="whitegrid", palette="pastel", color_codes=True)
```

```python
# Load the example tips dataset
tips = sns.load_dataset("tips")
tips.info()
```

```python
tips.sample()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
sns.violinplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
tips.sample()
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", split=True, data=tips)
```

```python
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(x="day", y="total_bill", hue="smoker",
               split=True, inner="quartile",
               data=tips)
sns.despine(left=True)
```
