## Univariate
* When there is only one variable
* Univariable - 1 variable (one independent variable)
* Univariate - univariate (one dependent variable)
* Source: [Independent Variables and Dependent Variables - Wikipedia, Encyclopedia of All of Us](https://en.wikipedia.org/wiki/%EB%8F%85%EB%A6%BD%EB%B3%80 %EC%88%98%EC%99%80_%EC%A2%85%EC%86%8D%EB%B3%80%EC%88%98)
    * Variables used in experiments or modeling can be classified as independent variables, dependent variables, or others. The independent variable represents the input value or cause, and the dependent variable represents the effect or effect.
    * Also called feature in machine learning or pattern recognition. Also referred to simply as an input variable.
    * Independent variables are also called explanatory variables. It is preferred only when the quantity of what is treated as an independent variable is not statistically independent.

## Frequency distribution
Source: [Frequency Distribution - Wikipedia, Encyclopedia of All of Us](https://en.wikipedia.org/wiki/%EB%8F%84%EC%88%98_%EB%B6%84%ED%8F %AC)

A frequency distribution is a list, table, or graph that shows the distribution of various outputs in a sample. Each entry in the table contains the frequency or number of occurrences of a value within a particular group or period, and in this way the table summarizes the distribution of sample values.

```python
# 라이브러리를 로드합니다.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

plt.style.use('seaborn')
```

```python
# tips 데이터셋을 가져옵니다.
tips = sns.load_dataset("tips")
tips.shape
```

```python
# 데이터프레임의 info를 봅니다.
# 이 때, 결측치가 있는지, 데이터의 타입이 무엇인지, 
# 데이터의 행과 열의 갯수는 몇 개인지, 메모리 사이즈는 어느정도인지를 봅니다.

tips.info()
```

### 시각화의 안 좋은 예

```python
# 기존 단변량 범주형 데이터에 적용했던 countplot 을 단변량 수치 데이터에 적용해 봅니다.
tips["tip"]
```

```python
sns.countplot(data=tips, x="tip")
```

```python
# countplot은 등장하는 모든 값에 대해 빈도수를 세게 됩니다.
# pandas 코드로 표현하면 value_counts 로 셀 수 있습니다.
tips["tip"].nunique()
```

### 데이터 타입에 맞는 시각화 

```python
# distplot 으로 시각화를 합니다. 
# 히스토그램과 kdeplot을 함께 그립니다.
sns.distplot(tips["tip"], rug=True)
```

```python
# bin 의 갯수를 조절해 봅니다. bin 1개 일 때

sns.distplot(tips["tip"], bins=1)
```

```python
# bin 의 갯수를 조절해 봅니다. bin 10개 일 때

sns.distplot(tips["tip"], bins=10)
```

```python
# kde = False 옵션으로 그려봅니다.
sns.distplot(tips["tip"], kde=False)
```

```python
# hist = False 옵션으로 그려봅니다.
sns.distplot(tips["tip"], hist=False, rug=True)
```

```python
# kdeplot 으로 그리기 
# kde는 KDE(kernel density estimation)의 약자
sns.kdeplot(tips["tip"], shade=True)
```

```python

sns.kdeplot(tips["tip"], cumulative=True, shade=True)
```

```python

sns.kdeplot(tips["tip"], shade=True, vertical=True)
```
