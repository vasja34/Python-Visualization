# Scatterplot with categorical variables
* https://seaborn.pydata.org/examples/scatterplot_categorical.html

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

plt.style.use('seaborn')
sns.set(style="whitegrid", palette="muted")
```

```python
# Load the example iris dataset
iris = sns.load_dataset("iris")
iris.sample()
```

```python
iris.info()
```

```python
# "Melt" the dataset to "long-form" or "tidy" representation
iris = pd.melt(iris, "species", var_name="measurement")
iris
```

```python
sns.scatterplot(x="measurement",  y="value",  hue="species", data=iris)
```

```python
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="measurement", y="value", hue="species",
              palette=["r", "c", "y"], data=iris)
```

```python
tips = sns.load_dataset("tips")
tips.sample()
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", inner="point", split=True, data=tips)
sns.swarmplot(x="day", y="total_bill", hue="smoker", data=tips)
```
