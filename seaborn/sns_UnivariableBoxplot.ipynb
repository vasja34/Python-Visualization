## Univariate
* When there is only one variable
* Univariable - 1 variable (one independent variable)
* Univariate - univariate (one dependent variable)
    * Variables used in experiments or modeling can be classified as independent variables, dependent variables, or others. 
    The independent variable represents the input value or cause, and the dependent variable represents the effect or effect.
    * Also called feature in machine learning or pattern recognition. Also referred to simply as an input variable.
    * Independent variables are also called explanatory variables. It is preferred only when the quantity of what is treated as an independent variable is not statistically independent.

### box-and-whisker plot | diagram
* It is not drawn using raw data as it is, but is drawn with five summary figures, which are statistics obtained from the data.
* The five summary figures refer to the five figures below that provide information about data in descriptive statistics.


1. Minimum
1. First quartile
1. Second quartile ( ), that is, the median
1. 3rd quartile ( )
1. Maximum

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(15, 8))

sns.distplot(tips["tip"], ax=axes[0])
sns.boxplot(tips["tip"], ax=axes[1])
sns.boxenplot(tips["tip"], ax=axes[2])
sns.violinplot(tips["tip"], ax=axes[3])
sns.swarmplot(tips["tip"], ax=axes[4])
```

## Finding outliers
* IQR = Q3 - Q1
* outlier
     * Q3 + 1.5 (IQR)
     * Q1 - 1.5 (IQR)

```python
desc = tips["tip"].describe()
desc
```

```python
IQR = desc["75%"] - desc["25%"]
IQR
```

```python
IQR = tips["tip"].quantile(0.75) - tips["tip"].quantile(0.25)
IQR
```

```python
IQR_15 = IQR * 1.5
IQR_15
```

```python
desc["75%"] + IQR_15
```

```python
desc["25%"] - IQR_15
```
