```python
import seaborn as sns
%matplotlib inline

sns.set(style="ticks")

# Load the example dataset for Anscombe's quartet
# Import the built-in dataset. For a description of this dataset, see the Wikipedia link above.
df = sns.load_dataset("anscombe")
```

```python
df.shape
```

```python
# If you look at the summary of the data, the mean and standard deviation are the same.
df.describe()
```

```python
# If you look at the summary of the data, the mean and standard deviation are the same.
df[df["dataset"] == "I"].describe()
```

```python
df[df["dataset"] == "II"].describe()
```

```python
df[df["dataset"] == "III"].describe()
```

```python
df[df["dataset"] == "IV"].describe()
```

```python
df.groupby(["dataset"]).describe()
```

## Pearson's correlation coefficient
* The Pearson correlation coefficient is the covariance of two variables divided by the product of the standard deviation. [Source: Wikipedia]( https://en.wikipedia.org/wiki/Pearson%27s_correlation_coefficient)

```python
# The correlation coefficient is the same up to three decimal places.
df[df["dataset"] == "I"].corr()
```

```python
df[df["dataset"] == "II"].corr()
```

```python
df[df["dataset"] == "III"].corr()
```

```python
df[df["dataset"] == "IV"].corr()
```

```python
# You can view the document by pressing Shift + Tab within parentheses () of the method.
sns.lmplot(x="x", y="y", data=df)
```

```python
# hue expresses the graph in different colors depending on the type of data.
# Here, the regression lines are all the same, so the lines are expressed as one, but they appear to overlap.
sns.lmplot(x="x", y="y", data=df, hue="dataset")
```

```python
# regplot has almost the same functionality as lmplot, but without the hue option.
sns.regplot(x="x", y="y", data=df)
```

```python
# lmplot
sns.lmplot(x="x", y="y", data=df, hue="dataset", col="dataset")
```

```python
# ci == Size of the confidence interval for the regression estimate
# ci is the confidence interval and has a value from 0 to 100.
# If it's 95, it means there's a 95% chance that the number will be in that range when you sample the data.
# Here, the confidence interval drawn above is excluded.
sns.lmplot(x="x", y="y", data=df, hue="dataset", col="dataset", ci=None)
```

```python
# height means the height of the graph
# In scatter_kws, s is the point size and alpha is the transparency.
sns.lmplot(x="x", y="y", data=df, hue="dataset", col="dataset", 
           col_wrap=2, ci=None, palette="muted", height=4, scatter_kws={"s": 70, "alpha": 0.8})
```

```python
# Show the results of a linear regression within each dataset
# seaborn In order to understand the graph of the example on the site, I have drawn the options one by one above.
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```
