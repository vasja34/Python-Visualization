# Scatter Plot with Matplotlib in Python

## What is Scatter Plot?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

plt.style.use('seaborn')
# plt.rcParams["figure.figsize"] = (11, 8)
```

```python
iris = pd.read_csv("../datasets/iris.csv")
```

```python
iris.head()
```

## Scatter Plot with Plot Method

```python
plt.plot(iris.SepalLength,iris.SepalWidth, "o")
```

```python
plt.plot(iris.SepalLength,iris.SepalWidth, "+")
```

```python
plt.plot(iris.SepalLength,iris.SepalWidth, "v")
```

```python
plt.plot(iris.SepalLength,iris.SepalWidth, "rv")
```

```python
plt.plot(iris.SepalLength,iris.SepalWidth, 
         "rv", markersize=10)
```

## Scatter Plot with Scatter Method

```python
plt.scatter(iris.SepalLength,iris.SepalWidth)
```

```python
plt.scatter(iris.SepalLength,
            iris.SepalWidth,s=75)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=75, c="m")
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=75, c="m", marker="v")
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=75, c="m", 
            edgecolor="black", 
            linewidth=1)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth,
            s=75, c="m", 
            edgecolor="black", 
            linewidth=1, 
            alpha=0.5)
```

```python
color=np.random.randint(0,10,150)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=100, 
            c=color, 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=100, 
            c=color, 
            cmap="Reds", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
cbar.set_label("Levels")
```

```python
sizes=np.random.randint(50,200,150)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("SepalWidth")
plt.ylabel("petal_length")
```

```python
plt.scatter(iris.SepalLength,iris.SepalWidth, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("SepalWidth")
plt.ylabel("petal_length")
plt.title("İris Sepal Length & Sepal Width ")
```
